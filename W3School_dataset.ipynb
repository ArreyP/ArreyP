```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"], format="%d/%m/%Y") #converts the date stored in df above from String to Datetime
df.info()
```

```python
df["ProductValue"] = df["Price"]*df["Quantity"]
print(df["ProductValue"])
```

```python
for ShipperID in df:
  sns.displot(data=df, x=ShipperID)
```

```python
#overall sales
sales_by_date = df.groupby('OrderDate')['ProductValue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_by_date, x='OrderDate', y='ProductValue')
plt.title('Total Sales Over 3-Years')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.tight_layout()
plt.show()
```

```python
#Top 17 Products by Sales
top_products = df.groupby('ProductName')['ProductValue'].sum().nlargest(17).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='ProductValue', y='ProductName', palette='viridis')
plt.title('Top 10 Products by Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
#top countries
country_sales = df.groupby('Country')['ProductValue'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=country_sales, x='ProductValue', y='Country', palette='magma')
plt.title('Top 10 Countries by Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#Sales by continents
country_sales = df.groupby('continent')['ProductValue'].sum().sort_values(ascending=False).head().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=country_sales, x='ProductValue', y='continent', palette='magma')
plt.title('Top Sales by continents')
plt.xlabel('Sales ($)')
plt.ylabel('continent')
plt.tight_layout()
plt.show()
```

```python
#Sales by Years
df['OrderYear'] = df['OrderDate'].dt.to_period('Y').astype(str)

monthly_sales = df.groupby('OrderYear')['ProductValue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='OrderYear', y='ProductValue', marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#Sales by months
df['OrderMonth'] = df['OrderDate'].dt.to_period('M').astype(str)

monthly_sales = df.groupby('OrderMonth')['ProductValue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='OrderMonth', y='ProductValue', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Make sure OrderDate is datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

# Optional: Clean up missing ProductValue
df = df[df['ProductValue'].notna()]
```

```python
#KPI metrics
total_sales = df['ProductValue'].sum()
avg_order_value = df.groupby('OrderID')['ProductValue'].sum().mean()
total_orders = df['OrderID'].nunique()
top_product = df.groupby('ProductName')['ProductValue'].sum().idxmax()

print(f"Total Sales: £{total_sales:,.2f}")
print(f"Average Order Value: £{avg_order_value:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Top Product: {top_product}")
```

```python
# Sales over time
fig1 = px.line(df.groupby('OrderDate')['ProductValue'].sum().reset_index(),
               x='OrderDate', y='ProductValue', title='Sales Over Time')

# Top products
top_products = df.groupby('ProductName')['ProductValue'].sum().nlargest(10).reset_index()
fig2 = px.bar(top_products, x='ProductValue', y='ProductName', orientation='h', title='Top 10 Products')

# Sales by country
top_countries = df.groupby('Country')['ProductValue'].sum().nlargest(10).reset_index()
fig3 = px.bar(top_countries, x='ProductValue', y='Country', orientation='h', title='Sales by Country')

# Monthly trend
df['OrderMonth'] = df['OrderDate'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('OrderMonth')['ProductValue'].sum().reset_index()
fig4 = px.line(monthly_sales, x='OrderMonth', y='ProductValue', title='Monthly Sales Trend')

# Display all
fig1.show()
fig2.show()
fig3.show()
fig4.show()
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=3, specs=[[{"type":"indicator"}]*3])

fig.add_trace(go.Indicator(
    mode="number",
    value=total_sales,
    title="Total Sales (£)"
), row=1, col=1)

fig.add_trace(go.Indicator(
    mode="number",
    value=avg_order_value,
    title="Avg Order Value (£)"
), row=1, col=2)

fig.add_trace(go.Indicator(
    mode="number",
    value=total_orders,
    title="Total Orders"
), row=1, col=3)

fig.update_layout(title="Key Performance Indicators", height=200)
fig.show()
```
